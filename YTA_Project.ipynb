```python
import pandas as pd

# Assuming you have uploaded 'electricity_cost_dataset.csv' to the Colab environment
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/electricity_cost_dataset.csv')

# Display the first 5 rows and information about the DataFrame
print('DataFrame head:')
display(df.head())
print('\nDataFrame info:')
display(df.info())
print('\nDataFrame description:')
display(df.describe())
```

```python
# Rename columns for clarity based on the dataset description
df = df.rename(columns={
    'site area': 'Site Area (square meters)',
    'structure type': 'Structure Type',
    'water consumption': 'Water Consumption (liters/day)',
    'recycling rate': 'Recycling Rate (%)',
    'utilisation rate': 'Utilisation Rate (%)',
    'air qality index': 'Air Quality Index (AQI)', # Correcting typo and adding detail
    'issue reolution time': 'Issue Resolution Time (hours)', # Correcting typo and adding detail
    'resident count': 'Resident Count (number of people)',
    'electricity cost': 'Electricity Cost (USD/month)'
})

print('DataFrame head with updated column names:')
display(df.head())
print('\nDataFrame info with updated column names:')
display(df.info())
```

### Step 2: Clean the Data

#### 2.1 Missing Values
First, let's check for any missing values in our DataFrame. Even though `df.info()` shows no nulls, it's good practice to explicitly check using `isnull().sum()`.

```python
# Check for missing values
missing_values = df.isnull().sum()
print('Missing values in each column:')
display(missing_values[missing_values > 0])

# Fill missing numerical values with the mean
# Note: From df.info(), it appears there are no missing values in this dataset, but this code demonstrates the fix.
# This operation will only affect numerical columns if they had NaNs.
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    if df[col].isnull().any(): # Only apply if there are actually missing values
        df[col].fillna(df[col].mean(), inplace=True)
        print(f"Filled missing values in '{col}' with its mean.")

if missing_values[missing_values > 0].empty:
    print("\nNo missing values found in the dataset, so no values were filled.")
```

#### 2.2 Fix Inconsistent Categorical Data
Next, we'll check the 'Structure Type' column for any inconsistencies, such as variations in how a single category might be represented (e.g., "AC" vs. "air conditioner" as mentioned in your example). We'll inspect the unique values in this column.

```python
# Check unique values in 'Structure Type' for inconsistencies
unique_structure_types = df['Structure Type'].unique()
print('Unique Structure Types:')
print(unique_structure_types)

# Based on the dataset description, the expected values are 'Residential', 'Commercial', 'Mixed-use', and 'Industrial'.
# If there were inconsistencies (e.g., 'Residntl' instead of 'Residential'), we would use .replace():
# df['Structure Type'] = df['Structure Type'].replace({'Residntl': 'Residential'})
# For this dataset, the values appear consistent based on your description.
```

#### 2.3 Remove Outliers
Your instruction mentions `df = df[df['power'] < 10000]` for outlier removal. However, a column named `power` does not exist in our DataFrame.

I will check for the `power` column. If it's not found, I will proceed with a common method for outlier removal on the `Electricity Cost (USD/month)` column, for example, by removing values above the 99th percentile to handle potential extreme costs.

```python
if 'power' in df.columns:
    print("'power' column found. Applying outlier removal based on 'power' < 10000.")
    original_rows = len(df)
    df = df[df['power'] < 10000]
    print(f"Removed {original_rows - len(df)} rows as outliers from 'power' column.")
else:
    print("The 'power' column was not found in the DataFrame.")
    print("Applying outlier removal to 'Electricity Cost (USD/month)' by keeping values below the 99th percentile.")

    original_rows = len(df)
    # Calculate the 99th percentile of 'Electricity Cost (USD/month)'
    cost_upper_bound = df['Electricity Cost (USD/month)'].quantile(0.99)

    # Remove outliers
    df = df[df['Electricity Cost (USD/month)'] < cost_upper_bound]
    print(f"Removed {original_rows - len(df)} rows as outliers from 'Electricity Cost (USD/month)' (above {cost_upper_bound:.2f} USD).")

print('\nDataFrame head after outlier removal:')
display(df.head())
print('\nNew DataFrame info after cleaning:')
display(df.info())
```

### Step 3: Feature Engineering and Selection

#### 3.1 One-Hot Encoding for Categorical Features
We need to convert the 'Structure Type' categorical column into a numerical format that machine learning models can understand. One-hot encoding is a suitable method for this.

```python
# Apply one-hot encoding to 'Structure Type'
df_encoded = pd.get_dummies(df, columns=['Structure Type'], drop_first=True)

print('DataFrame head after one-hot encoding:')
display(df_encoded.head())
print('\nDataFrame info after one-hot encoding:')
display(df_encoded.info())
```

### Step 4: Select Features (X) and Target (y)

Now we will separate our dataset into features (X) and the target variable (y). The target variable is 'Electricity Cost (USD/month)'.

```python
# Define features (X) and target (y)
X = df_encoded.drop('Electricity Cost (USD/month)', axis=1)
y = df_encoded['Electricity Cost (USD/month)']

print('Features (X) head:')
display(X.head())
print('\nTarget (y) head:')
display(y.head())
```

### Step 5: Split the Data

Now, we'll split our dataset into training and testing sets. This is a crucial step for evaluating our machine learning model, as it allows us to train the model on one portion of the data and test its performance on unseen data.

**VERY IMPORTANT ⚠️**

Train ≠ Test (never mix them)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Added random_state for reproducibility

print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

### Step 6: Preprocessing

#### 6.1 Normalize Data
We will use `StandardScaler` to normalize our numerical features. Normalization (or standardization) is a common preprocessing step that scales the features such that they have a mean of 0 and a standard deviation of 1. This helps to prevent features with larger values from dominating the learning process and can improve the performance of many machine learning algorithms.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print('X_train_scaled shape:', X_train_scaled.shape)
print('X_test_scaled shape:', X_test_scaled.shape)

# Display a few scaled values to demonstrate
print('\nFirst 5 rows of X_train_scaled:')
print(X_train_scaled[:5])
print('\nFirst 5 rows of X_test_scaled:')
print(X_test_scaled[:5])
```

### Step 7: Train your first model (Baseline)

We'll start with a simple Linear Regression model as a baseline. This helps us establish a basic performance metric before moving on to more complex models.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
# Train the model using the scaled training data
model.fit(X_train_scaled, y_train)

print("Linear Regression model trained successfully on scaled data!")
```

It's always a good practice to start with a simple model like Linear Regression as a baseline. This helps in understanding the basic relationships in the data and provides a benchmark against which more complex models can be compared.

### Step 8: Evaluate the model

Now, let's evaluate the performance of our baseline Linear Regression model using `Mean Squared Error (MSE)`. MSE is a common metric for regression tasks, measuring the average of the squares of the errors—that is, the average squared difference between the estimated values and the actual value.

```python
from sklearn.metrics import mean_squared_error

# Make predictions on the scaled test data
pred = model.predict(X_test_scaled)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Predictions on original (unscaled) X_train and X_test for evaluation metrics
# We need to predict using the scaled X_train_scaled and X_test_scaled since the model was trained on scaled data
y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test_scaled)

# Metrics
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_test)

print("Model Performance:")
print(f"Train R²: {train_r2:.4f}")
print(f"Test R²: {test_r2:.4f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```

### Step 9: Visualize the Results

Let's visualize the model's performance by comparing the predicted electricity costs with the actual costs for the test set. This will help us understand how well our model is performing visually.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot of actual vs. predicted values
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred_test, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2) # Line for perfect prediction
plt.xlabel('Actual Electricity Cost (USD/month)')
plt.ylabel('Predicted Electricity Cost (USD/month)')
plt.title('Actual vs. Predicted Electricity Cost (Test Set)')
plt.grid(True)
plt.show()
```

#### Residual Plot

A residual plot helps to visualize the errors of the predictions. Ideally, residuals should be randomly scattered around zero, indicating that the model is capturing the underlying patterns in the data without any systematic bias.

```python
# Calculate residuals
residuals = y_test - y_pred_test

# Create a residual plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred_test, y=residuals, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--', lw=2)
plt.xlabel('Predicted Electricity Cost (USD/month)')
plt.ylabel('Residuals (Actual - Predicted)')
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```

### Step 10: Train a Decision Tree Regressor Model on Original Data

### Using `model.score()` for Regression Models

For scikit-learn regression models, the `score()` method returns the **coefficient of determination, R²** of the prediction. This metric indicates how well the model's predictions explain the variability of the target variable. A value of 1.0 indicates a perfect fit, while 0.0 indicates that the model explains no variability in the target. It's an important metric to understand the model's performance on unseen data.

```python
# Score for the Linear Regression model
linear_regression_score = model.score(X_test_scaled, y_test)
print(f"Linear Regression Model (R² score): {linear_regression_score:.4f}")

# Score for the Decision Tree Regressor model
decision_tree_score = dt_model.score(X_test_scaled, y_test)
print(f"Decision Tree Regressor Model (R² score): {decision_tree_score:.4f}")
```

### Step 18: Compare All Models

Let's compare the performance of all trained models (Linear Regression, Decision Tree, Random Forest, XGBoost, LightGBM, and CatBoost) based on their R², RMSE, and MAE scores on the test set.

```python
import pandas as pd

# Store metrics in a dictionary
model_performance = {
    'Linear Regression': {'R²': test_r2, 'RMSE': rmse, 'MAE': mae},
    'Decision Tree': {'R²': dt_test_r2, 'RMSE': dt_rmse, 'MAE': dt_mae},
    'Random Forest': {'R²': rf_test_r2, 'RMSE': rf_rmse, 'MAE': rf_mae},
    'XGBoost': {'R²': xgb_test_r2, 'RMSE': xgb_rmse, 'MAE': xgb_mae},
    'LightGBM': {'R²': lgbm_test_r2, 'RMSE': lgbm_rmse, 'MAE': lgbm_mae},
    'CatBoost': {'R²': cat_test_r2, 'RMSE': cat_rmse, 'MAE': cat_mae}
}

# Convert to DataFrame for easy comparison
performance_df = pd.DataFrame(model_performance).T

print("\n--- Model Performance Comparison (Test Set) ---")
display(performance_df.sort_values(by='R²', ascending=False))

print("\n--- Best Model for each Metric ---")
# Best R² (higher is better)
best_r2_model = performance_df['R²'].idxmax()
best_r2_score = performance_df['R²'].max()
print(f"Best R²: {best_r2_model} with R² = {best_r2_score:.4f}")

# Best RMSE (lower is better)
best_rmse_model = performance_df['RMSE'].idxmin()
best_rmse_score = performance_df['RMSE'].min()
print(f"Best RMSE: {best_rmse_model} with RMSE = {best_rmse_score:.2f}")

# Best MAE (lower is better)
best_mae_model = performance_df['MAE'].idxmin()
best_mae_score = performance_df['MAE'].min()
print(f"Best MAE: {best_mae_model} with MAE = {best_mae_score:.2f}")

print("\n--- Best Model Overall (considering R² as primary metric) ---")
# For overall best, R² is usually the primary indicator for how well the model explains variance.
# If multiple models have very similar R², you might look at RMSE/MAE as tie-breakers.
overall_best_model = performance_df['R²'].idxmax()
print(f"The overall best model is: {overall_best_model}")
print(f"With R²: {performance_df.loc[overall_best_model, 'R²']:.4f}, RMSE: {performance_df.loc[overall_best_model, 'RMSE']:.2f}, MAE: {performance_df.loc[overall_best_model, 'MAE']:.2f}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, LeakyReLU
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Define the MLP model with LeakyReLU activation
mlp_leakyrelu_model = Sequential([
    Dense(128, input_shape=(X_train_scaled.shape[1],)),
    BatchNormalization(),
    LeakyReLU(alpha=0.1),
    Dense(64),
    BatchNormalization(),
    LeakyReLU(alpha=0.1),
    Dense(32),
    BatchNormalization(),
    LeakyReLU(alpha=0.1),
    Dense(1) # Output layer for regression
])

# Compile the model
mlp_leakyrelu_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Train the model
print("Training MLP (LeakyReLU) model... This might take some time.")
history_leakyrelu = mlp_leakyrelu_model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)
print("MLP (LeakyReLU) model trained successfully!")

# Make predictions
y_mlp_leakyrelu_pred_train = mlp_leakyrelu_model.predict(X_train_scaled).flatten()
y_mlp_leakyrelu_pred_test = mlp_leakyrelu_model.predict(X_test_scaled).flatten()

# Calculate metrics
mlp_leakyrelu_train_r2 = r2_score(y_train, y_mlp_leakyrelu_pred_train)
mlp_leakyrelu_test_r2 = r2_score(y_test, y_mlp_leakyrelu_pred_test)
mlp_leakyrelu_mse = mean_squared_error(y_test, y_mlp_leakyrelu_pred_test)
mlp_leakyrelu_rmse = np.sqrt(mlp_leakyrelu_mse)
mlp_leakyrelu_mae = mean_absolute_error(y_test, y_mlp_leakyrelu_pred_test)

print("\nDeep Neural Network (MLP) with LeakyReLU Model Performance:")
print(f"Train R²: {mlp_leakyrelu_train_r2:.4f}")
print(f"Test R²: {mlp_leakyrelu_test_r2:.4f}")
print(f"RMSE: {mlp_leakyrelu_rmse:.2f}")
print(f"MAE: {mlp_leakyrelu_mae:.2f}")
```

### Step 33: Train a Deep Neural Network (LSTM Model)

We will now implement a Long Short-Term Memory (LSTM) network. While LSTMs are typically used for sequential data, we can adapt them for tabular data by treating each sample as a sequence of one timestep.

First, we need to reshape our input data `X_train_scaled` and `X_test_scaled` to fit the LSTM input format, which is `(samples, timesteps, features)`.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, BatchNormalization
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Reshape data for LSTM: (samples, timesteps, features)
# For tabular data, we can treat each sample as a single timestep
X_train_reshaped = X_train_scaled.reshape(X_train_scaled.shape[0], 1, X_train_scaled.shape[1])
X_test_reshaped = X_test_scaled.reshape(X_test_scaled.shape[0], 1, X_test_scaled.shape[1])

print(f"X_train_reshaped shape: {X_train_reshaped.shape}")
print(f"X_test_reshaped shape: {X_test_reshaped.shape}")

# Define the LSTM model
lstm_model = Sequential([
    LSTM(128, activation='relu', input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2]), return_sequences=True),
    BatchNormalization(),
    LSTM(64, activation='relu', return_sequences=False),
    BatchNormalization(),
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dense(1) # Output layer for regression
])

# Compile the model
lstm_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Train the model
print("Training LSTM model... This might take some time.")
history_lstm = lstm_model.fit(X_train_reshaped, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)
print("LSTM model trained successfully!")

# Make predictions
y_lstm_pred_train = lstm_model.predict(X_train_reshaped).flatten()
y_lstm_pred_test = lstm_model.predict(X_test_reshaped).flatten()

# Calculate metrics
lstm_train_r2 = r2_score(y_train, y_lstm_pred_train)
lstm_test_r2 = r2_score(y_test, y_lstm_pred_test)
lstm_mse = mean_squared_error(y_test, y_lstm_pred_test)
lstm_rmse = np.sqrt(lstm_mse)
lstm_mae = mean_absolute_error(y_test, y_lstm_pred_test)

print("\nDeep Neural Network (LSTM) Model Performance:")
print(f"Train R²: {lstm_train_r2:.4f}")
print(f"Test R²: {lstm_test_r2:.4f}")
print(f"RMSE: {lstm_rmse:.2f}")
print(f"MAE: {lstm_mae:.2f}")
```
